```python
def get_total_marks():
    marks = [85, 90, 78]  # example marks
    return sum(marks)

def assign_grade(total_marks):
    if total_marks >= 250:
        return "A"
    elif total_marks >= 200:
        return "B"
    else:
        return "C"

# Driver Code
total = get_total_marks()
print("Grade:", assign_grade(total))
```

```python
def calculate_total_bill(p1, p2, p3):
    return p1 + p2 + p3

def apply_discount(bill_amount):
    if bill_amount > 5000:
        discount = bill_amount * 0.10
    else:
        discount = bill_amount * 0.05
    print("Final Amount:", bill_amount - discount)

# Driver Code
bill = calculate_total_bill(2000, 1500, 1800)
apply_discount(bill)
```

```python
def calculate_tax(income):
    if income <= 200000:
        return 0
    elif income <= 500000:
        return income * 0.05
    else:
        return income * 0.10

# Driver Code
print("Tax:", calculate_tax(600000))
```

```python
def calculate_zakat(weight):
    price_per_gram = 8000
    return weight * price_per_gram * 0.025

# Driver Code
print("Gold Zakat:", calculate_zakat(50))
```

```python
def calculate_profit_loss(cost_price, selling_price):
    profit_loss = ((selling_price - cost_price) / cost_price) * 100
    return profit_loss

# Driver Code
result = calculate_profit_loss(1000, 1200)
print("Profit Percentage:" if result > 0 else "Loss Percentage:", abs(result))
```

```python
def calculate_discounted_price(price, discount_percent):
    return price - (price * discount_percent / 100)

# Driver Code
print("Discounted Price:", calculate_discounted_price(2000, 15))
```

```python
def calculate_basic_bill(units):
    if units <= 100:
        return units * 5
    elif units <= 200:
        return (100 * 5) + ((units - 100) * 7)
    else:
        return (100 * 5) + (100 * 7) + ((units - 200) * 10)

def calculate_total_bill(basic_bill):
    if basic_bill > 1500:
        return basic_bill + (basic_bill * 0.10)
    return basic_bill

# Driver Code
units = 250
basic = calculate_basic_bill(units)
total = calculate_total_bill(basic)
print("Total Bill:", total)
```

```python
string = "Python Programming"
vowels = "aeiouAEIOU"
v_count = sum(1 for c in string if c in vowels)
c_count = sum(1 for c in string if c.isalpha() and c not in vowels)
print("Vowels:", v_count, "Consonants:", c_count)
```

```python
import string

sentence = "Hello, world! Python is great."
words = sentence.split()
print("Reversed:", " ".join(words[::-1]))
print("Capitalized:", " ".join(w.capitalize() for w in words))
print("Without Punctuation:", sentence.translate(str.maketrans('', '', string.punctuation)))
```

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 30, 'c': 40}
merged = dict1.copy()
for k, v in dict2.items():
    merged[k] = merged.get(k, 0) + v
print("Merged Dict:", merged)
```

```python
data = {'a': 10, 'b': 20, 'c': 10}
values_seen = set()
unique = {}
for k, v in data.items():
    if v not in values_seen:
        unique[k] = v
        values_seen.add(v)
print("Unique Dict:", unique)
```

```python
students = {
    'Ali': {'Math': 80, 'Science': 70},
    'Sara': {'Math': 90, 'Science': 85}
}
for name, subjects in students.items():
    avg = sum(subjects.values()) / len(subjects)
    print(f"{name} Average:", avg)
```
